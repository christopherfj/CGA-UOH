```python
def perfecto(n):        
    divisor = 1
    suma = 0
    while divisor<n:
        if (n%divisor)==0:
            suma+=divisor
        divisor+=1
    if suma==n:
        print(n,'es un número perfecto')
    else:
        print(n,'no es un número perfecto')  

perfecto(12)
perfecto(28)
```
